```python
michael_files=FileUtility.recursive_glob('/mounts/data/proj/asgari/superparallelproj/paralleltext/bibles/corpus/','*.txt')
```

```python
lang_length=dict()
for x in tqdm.tqdm(michael_files):
    l=len(FileUtility.load_list(x))
    x=x.split('/')[-1][0:3]
    if x not in lang_length:
        lang_length[x]=0
    lang_length[x]=max(l,lang_length[x])
```

```python
df=pd.read_table('/mounts/data/proj/asgari/final_proj/000_datasets/testbib/API/reports/final_rep.tsv')
```

```python
my_length=dict()
for x ,y in df.groupby('language_iso')['verses'].apply(list).to_dict().items():
    my_length[x]=max(y)
```

```python
a=set(lang_length.keys())
for x in set(my_length.keys()):
    a.add(x)
```

```python
report={'iso':[],'michael':[],'ours':[]}

for x in a:
    report['iso'].append(x)
    if x in my_length:
        report['ours'].append(my_length[x])
    else:
        report['ours'].append(0)
    if x in lang_length:
        report['michael'].append(lang_length[x]-15)
    else:
        report['michael'].append(0)

        
    
```

```python
df=pd.DataFrame(report)
df=df.set_index('iso')
df.to_csv('../meta/comparison.tsv',sep='\t')
```
